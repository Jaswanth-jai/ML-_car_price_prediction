# Model

```python
car = pd.read_csv('Cleaned car.csv')
car
```

```python
car.columns
```

```python
p = car.drop('company' , axis=1)
p.columns
```

```python
x = p.drop('Price', axis=1)

y = car['Price']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from  sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score , accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import  make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor

```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name', 'fuel_type']])
```

```python


colmn_trans = make_column_transformer(
    (OneHotEncoder(categories=ohe.categories_), ['name',  'fuel_type']),
    remainder='passthrough'
)
```

```python
lr = LinearRegression()
rnf = RandomForestRegressor()
```

```python
pipe1 = make_pipeline(colmn_trans, lr)
pipe2 = make_pipeline(colmn_trans, rnf)
```

```python
pipe1.fit(x_train, y_train)
pipe2.fit(x_train, y_train)
```

```python
y_pred1 = pipe1.predict(x_test)
y_pred2 = pipe2.predict(x_test)
```

```python
print(r2_score(y_test, y_pred1))
```

```python
print(r2_score(y_test, y_pred2))
```

```python
dt = {}
s = []
for i  in range(1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(colmn_trans, lr)
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    dt[i] = r2_score(y_test, y_pred)
    s.append(r2_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=pipe2, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Score:", grid_search.best_score_)
```

```python

final_pipe = make_pipeline(
    colmn_trans,
    RandomForestRegressor(
        max_depth=None,
        max_features='log2',
        min_samples_leaf=1,
        min_samples_split=2,
        n_estimators=100,
        random_state=42
    )
)

final_pipe.fit(x_train, y_train)
```

```python
new_pred2 = final_pipe.predict(x_test)
```
